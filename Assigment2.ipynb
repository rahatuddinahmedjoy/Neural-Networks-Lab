‚úÖ PART 1: Unique Dataset Design
üéØ My Labeling Rule (You can modify slightly)

A person is considered Fit (1) if:

Activity ‚â• 4 hours/week

Sleep between 6 and 8 hours

Water ‚â• 2 liters

Otherwise ‚Üí Unfit (0)

üìä Dataset (15 Data Points ‚Äì Balanced)

```python
# -------------------------------
# DATASET CREATION
# -------------------------------

# Each data point:
# [activity_hours_per_week, sleep_hours_per_day, water_intake_liters]

# Labeling Rule:
# Fit (1) if:
# activity >= 4 AND
# 6 <= sleep <= 8 AND
# water >= 2
# Otherwise Unfit (0)

features = [
    [5, 7, 2.5],   # Fit
    [6, 8, 2.0],   # Fit
    [4, 6, 2.2],   # Fit
    [3, 7, 2.5],   # Unfit (low activity)
    [5, 5, 2.5],   # Unfit (low sleep)
    [2, 6, 1.5],   # Unfit
    [7, 7, 3.0],   # Fit
    [1, 8, 2.0],   # Unfit
    [4, 7, 1.0],   # Unfit (low water)
    [6, 6, 2.8],   # Fit
    [3, 5, 1.8],   # Unfit
    [8, 7, 2.5],   # Fit
    [5, 9, 2.5],   # Unfit (too much sleep)
    [4, 6, 2.0],   # Fit
    [2, 4, 1.0]    # Unfit
]

labels = [1,1,1,0,0,0,1,0,0,1,0,1,0,1,0]
```

‚úÖ PART 2: Perceptron Code (From Scratch)

NO numpy. Only loops + lists.

```python
# -------------------------------
# PERCEPTRON INITIALIZATION
# -------------------------------

# Initial weights (chosen small to avoid large jumps)
w1 = 0.2
w2 = -0.1
w3 = 0.3

bias = 0.0

learning_rate = 0.01   # small for stable learning
epochs = 20


# -------------------------------
# ACTIVATION FUNCTION
# -------------------------------

def predict(x1, x2, x3):
    # Weighted sum
    weighted_sum = (w1 * x1) + (w2 * x2) + (w3 * x3) + bias

    # Step function with threshold 0
    if weighted_sum >= 0:
        return 1
    else:
        return 0


# -------------------------------
# TRAINING LOOP
# -------------------------------

for epoch in range(epochs):
    total_error = 0

    for i in range(len(features)):

        x1 = features[i][0]
        x2 = features[i][1]
        x3 = features[i][2]

        target = labels[i]

        output = predict(x1, x2, x3)

        error = target - output
        total_error += abs(error)

        # Perceptron update rule:
        # w = w + learning_rate * error * input

        w1 = w1 + learning_rate * error * x1
        w2 = w2 + learning_rate * error * x2
        w3 = w3 + learning_rate * error * x3
        bias = bias + learning_rate * error

    print("Epoch:", epoch+1)
    print("Total Error:", total_error)
    print("Weights:", w1, w2, w3)
    print("Bias:", bias)
    print("----------------------")
```

‚úÖ PART 3: User Input Testing

```python
# -------------------------------
# USER INPUT TESTING
# -------------------------------

print("\nEnter lifestyle details:")

a = float(input("Activity hours per week: "))
s = float(input("Sleep hours per day: "))
w = float(input("Water intake (liters): "))

result = predict(a, s, w)

if result == 1:
    print("This person is likely to be FIT based on the given lifestyle.")
else:
    print("This person is likely to be UNFIT based on the given lifestyle.")
```
